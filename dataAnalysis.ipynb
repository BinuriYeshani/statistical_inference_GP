```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid")
import seaborn as sns
import rte
```

```python

data_df = pd.read_csv("data/dataFile.csv", encoding ='UTF-8' ) 

#data_df = pd.read_csv("data/dataFile.csv", encoding ='0D80â€“0DFF' )  
print(data_df.shape)
print(data_df.sample(5))
```

```python
######### remove nan ##########

data_df = data_df.dropna()
print(data_df.shape)
```

```python

vote_or_not = data_df.loc[(data_df['vote_or_not'] == 'Yes') | (data_df['vote_or_not'] == 'No')]
vote_or_not_type = vote_or_not.groupby(['vote_or_not', 'vote_for']).size()
vote_or_not_type.unstack().plot(kind='bar', stacked=True)
```

```python

genders = data_df.loc[(data_df['Gender'] == 'Male') | (data_df['Gender'] == 'Female')]
gender_type = genders.groupby(['Gender', 'vote_for']).size()
gender_type.unstack().plot(kind='bar', stacked=True)
```

```python

past_vote_for = data_df.loc[(data_df['past_vote_for'] == 'Maithripala Sirisena') | (data_df['past_vote_for'] == 'Mahinda Rajapaksha')]
past_vote_for_type = genders.groupby(['past_vote_for', 'vote_for']).size()
past_vote_for_type.unstack().plot(kind='bar', stacked=True)
```

```python

ethnicity = data_df.loc[(data_df['ethnicity'] == 'Sinhala') | 
                        (data_df['ethnicity'] == 'Tamil') |
                        (data_df['ethnicity'] == 'Muslim') |
                        (data_df['ethnicity'] == 'Other')]

ethnicity_type = ethnicity.groupby(['ethnicity', 'vote_for']).size()

ethnicity_type.unstack().plot(kind='bar', stacked=True)
```

```python

highest_educational_level    = data_df.loc[(data_df['highest_educational_level'] == 'Grade5') | 
                                (data_df['highest_educational_level'] == 'OrdinaryLevel') |
                                (data_df['highest_educational_level'] == 'AdvancedLevel') |
                                (data_df['highest_educational_level'] == 'BachelorsDegreeHolder')|
                                (data_df['highest_educational_level'] == 'MastersorDoctoralDegreeHolder')|
                                (data_df['highest_educational_level'] == 'Other')]

education_type = highest_educational_level.groupby(['highest_educational_level', 'vote_for']).size()

education_type.unstack().plot(kind='bar', stacked=True)
```

```python

age = data_df.loc[(data_df['age'] == 'Below_18') | 
                (data_df['age'] == '18_35') |
                (data_df['age'] == '36_60') |
                (data_df['age'] == 'Above_60')]

age_type = age.groupby(['age', 'vote_for']).size()

age_type.unstack().plot(kind='bar', stacked=True)
```
